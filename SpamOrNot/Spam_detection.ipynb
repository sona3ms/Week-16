```python
data = pd.read_csv('/content/SPAM text message 20170820 - Data.csv')
data
```

```python
data['spam'] = data['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(data.Message, data.spam,test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
print(classification_report(y_test,y_pred))#class was imbalanced so we used claasification_report
```

```python
e = ['Hey sona, can we watch movie?',
     'upto 20% off on parking, exclusive offer for you. Dont miss this reward']
e_count = v.transform(e)
model.predict(e_count)
```

```python
#using pipeline
from sklearn.pipeline import Pipeline
clf = Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```
